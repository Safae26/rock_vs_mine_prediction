### <p style="color:blue"> Scenario : </p>

#### Consider a war going on between 2 countries.
#### A submarine of a country is going underwater to another country.
#### The enemy country have planted some mines (explosives) in the ocean
#### Of course, the ocean is full of rocks.
#### That is why the submarine needs to predict whether the object beneath is a mine or a rock.
#### This is done by the sonar in the submarine that uses sonar signal that sends sound signals and reviews switchbacks.
#### The signal is then processed to detect whether the object is a rock ðŸ’£ or a mine.

#### Data : In a laboratory setup, sonar is used to send and receive signals bounce back from a metal cylinder and some rocks.

#### Model/Algorithm : ðŸ“ˆ Logistic Regression (Supervised Learning), since we have a binary classification problem.

### <p style="color:blue"> Workflow : </p>

#### Collect Sonar Data --> Data Preprocessing --> Train Test Split --> ML Model (Logistic Regression) --> Trained Logistic Regression --> Prediction with new data (Metal Object or Rock withd probabilities)

### Let's import dependencies :

```python
import numpy as np # For Arrays
import pandas as pd # For loading DataFrames and Processing Data

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score # For evaluation
```

### Data Collection and Processing

```python
sonar_data = pd.read_csv("PATH_TO_DATASET", header=None)
```

```python
sonar_data.head()
```

```python
print("Number of rows and columns : ", sonar_data.shape)
```

```python
# Some Statistical definitions :

sonar_data.describe() # Statistical measures
```

```python
# Number of instances of "M" and "R"
sonar_data.value_counts(sonar_data.iloc[:, -1])
# Or : sonar_data[60].value_counts()
# Classes are almost equal, so the prediction will be good in term of accuracy
# Note : This is just a use case, that why we have this small number of data, or else we would need thounsandds of examples.
```

M --> Mine

R --> Rock

```python
sonar_data.groupby(60).mean()
```

```python
# Features and Target (Label) :
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting Data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
# stratify : we needs to have almost equal R and M in both Train and Test set
# random_state : to reproduce the code as it is
```

```python
print(f"Original data : {X.shape}.\nTrain set : {X_train.shape}.\nTest set : {X_test.shape}.")
```

```python
print(X_train)
```

```python
print(Y_train)
```

### Model Training

```python
# Loading the model
model= LogisticRegression()
```

```python
# Training it..
model.fit(X_train, Y_train)
```

```python
# Model Evaluation

X_train_pred = model.predict(X_train)
print("Accuracy on the training data : ", accuracy_score(X_train_pred, Y_train))
```

```python
X_test_pred = model.predict(X_test)
print("Accuracy on the testing data : ", accuracy_score(X_test_pred, Y_test))
```
